<h3>Las PROPS</h3>

<b>Intro</b> - Los componentes de React utilizan props para comunicarse entre sí. Cada componente padre puede enviar información a sus componentes hijos mediante el uso de props. Las props pueden parecerte similares a los atributos HTML, pero permiten pasar cualquier valor de JavaScript a través de ellas, como objetos, arrays y funciones.

<b>01. Comencemos con un arreglo</b> - Declaro un array con 4 posibles valores para el texto de un botón

```python
const listaBotones = ['Cancelar', 'Confirmar', 'Aceptar', 'Guardar']
```

<b>02. Componente</b> - Luego de declarar el arreglo, vamos a armar la estructura inicial de un componente. Recordar que se puede utilizar el atajo <i>rafc+tab</i> al inicializar un nuevo componente jsx

```python
export const EjemploProps = () => {
    return (
      <>
      //seleccionamos un item del array
        <button>{listaBotones[2]}</button>
      </>
    )
  }
```

<i>Tip. </i>Para insertar un contenido de js, ya sea una variable o una función, se hace mediante la inserción de JavaScript dentro de llave {nombreVariable}.
En código anterior, estamos seleccionando el item de la posición 2 del arreglo. Para nosotros sería el tercero, pero recordemos que el índice inicial siempre va a ser 0 en un array.

<b>03. Inserción del componente</b> - Coloquemos nuestro componente creado dentro de un contenedor padre (APP.JSX) para poder visualizarlo en el navegador.

```python
import EjemploProps from './ubicacion/EjemploProps'

export const App = () => {
  return (
    <>
    <EjemploProps />
    </>
  )
}
```

De esta manera ya podríamos vizualizar nuestro componente botón.

<b>04. Mapeo de un arreglo</b> - ¿Qué sucede si tenemos más botones? Hoy hablamos de botones como un componente sencillo, pero la regla que voy a explicar a continuación puede aplicarse a cualquier componente. Para ello vamos a recurrir a un método que nos permite transformar cada "hijo" en la estructura de datos.
<br>
    <i>Children.map(children, fn, thisArg?)</i>
<br>
<h5>Parámetros</h5>
<ul><li>children: El valor de la prop children que recibe por tu componente</li>
<li>fn: La función de mapeo, similar al callback del método map de un array. Se llamará con el hijo como primer argumento y su índice como segundo argumento. El índice empieza en 0 y se incrementa por cada llamada. Necesitas devolver un nodo de React de esta función. Puede ser un nodo vacío (null, undefined, o un booleano), un string, un número, un elemento de React, o un array de otros nodos de React.</li>
<li>opcional thisArg: El valor this con el que se debe llamar a la función fn. Si se omite, es undefined.</li></ul>



```python
const listaBotones = ['Confirmar', 'Cerrar', 'Aceptar', 'Guardar']

export const EjemploProps = () => {
  return (
    <>
    {listaBotones.map(boton =>
        <button>{boton}</button>)
        }
    </>

  )
}
```
